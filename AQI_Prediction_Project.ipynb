# **Scatter-Plott**

```python
# List of features to plot against AQI
features = ['CO', 'NO2', 'SO2', 'O3', 'PM2.5', 'PM10']

# Loop through each feature and plot
for feature in features:
    plt.figure(figsize=(6, 4))
    plt.scatter(AQI_df[feature], AQI_df['AQI'], alpha=0.3)
    plt.title(f'{feature} vs AQI')
    plt.xlabel(feature)
    plt.ylabel('AQI')
    plt.show()
```

# **Histogram-plot**

## **Encoding**

```python

#create dummy variables for the categorical variable state
AQI_df['Brasilia']=np.where(AQI_df['City']=='Brasilia',1,0)
AQI_df['Cairo']=np.where(AQI_df['City']=='Cairo',1,0)
AQI_df['Dubai']=np.where(AQI_df['City']=='Dubai',1,0)
AQI_df['London']=np.where(AQI_df['City']=='London',1,0)
AQI_df['New York']=np.where(AQI_df['City']=='New York',1,0)
AQI_df['Sydney']=np.where(AQI_df['City']=='Sydney',1,0)

#drop the original column state form the dataframe
AQI_df.drop(columns=['City'],axis=1,inplace=True)
```

```python
AQI_df
```

# **Impementing Linear Regression**

# **separating target variable from independent variable**

```python
#Dependent Variable
y=AQI_df['AQI']
#Independent Variable
x=AQI_df.drop('AQI',axis=1)
```

```python
y
```

```python
x
```

# **splitting the dataset into the Training set and Test set**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# **Standardization of data**

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train[0:10]
```

```python
x_test[0:10]
```

# **Applying linear regression model**

```python
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
regressor.intercept_
```

```python
regressor.coef_
```

# **Prediction**

```python
y_pred_train_linear=regressor.predict(x_train)
```

```python
y_train #actual value
```

```python
y_pred_train_linear #predicted value
```

```python
y_pred_test_linear=regressor.predict(x_test)
```

```python
y_test
```

```python
y_pred_test_linear
```

# **Train-Performance**

```python
mean_squared_error(y_train,y_pred_train_linear)
```

# **Test-Performance**
